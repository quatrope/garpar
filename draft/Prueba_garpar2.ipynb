```python

makerstable = gp.datasets.RissoLevyStable(n_jobs=8,random_state=42,alpha=1.6411,beta= -0.0126,mu=0.0005, sigma=0.005)
```

```python
mercado_stable = makerstable.make_portfolio(days=365, price= 150, window_size=5,stock_number=
                                            32,entropy=0.10)
```

```python
mercado_stable
```

```python
mercado_stable.plot.ogive()
```

```python
mercado_stable.plot.box()
```

```python
mercado_stable.plot.kde()
```

```python
optimizador = gp.optimize.Markowitz()
```

```python
optimizador
```

```python
portopt=optimizador.optimize(mercado_stable, 0.005)
```

```python
portopt.weights
```

```python
portopt.prune()
```

```python
portopt.to_hdf5("ejemplo.h5")
```
